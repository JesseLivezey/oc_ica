```python
rng = np.random.RandomState(20161206)
overcompleteness = 2
n_mixtures = 64
n_sources = n_mixtures*overcompleteness
n_iter = 1000
degeneracy_controls = ['2', '4', 'COULOMB_F', 'RANDOM_F']
initial_conditions = ['random']

W = np.full((len(degeneracy_controls), n_iter, n_sources,
             n_mixtures), np.nan)
W_0 = np.full((n_iter, n_sources, n_mixtures), np.nan)
for ii in range(n_iter):
    Wp, W_0p = analysis.evaluate_dgcs(initial_conditions, degeneracy_controls,
                                      n_sources, n_mixtures, rng=rng)
    W[:, ii] = Wp[0]
    W_0[ii] = W_0p
```

```python
min_angles = np.zeros((len(degeneracy_controls), n_iter))
costs = np.zeros((len(degeneracy_controls), n_iter))
for ii, dc in enumerate(degeneracy_controls):
    try:
        dc = int(dc)
        dc = 'L{}'.format(dc)
    except ValueError:
        pass
    model = ICA(n_mixtures, n_sources, lambd=0., degeneracy=dc)
    for jj in range(n_iter):
        min_angles[ii, jj] = analysis.compute_angles(W[ii, jj]).min()
        model.components_ = W[ii, jj]
        costs[ii, jj] = model.losses(np.zeros((n_mixtures, 1)))[0]
        
min_angles0 = np.zeros((len(degeneracy_controls), n_iter))
costs0 = np.zeros((len(degeneracy_controls), n_iter))
for ii, dc in enumerate(degeneracy_controls):
    try:
        dc = int(dc)
        dc = 'L{}'.format(dc)
    except ValueError:
        pass
    model = ICA(n_mixtures, n_sources, lambd=0., degeneracy=dc)
    for jj in range(n_iter):
        min_angles0[ii, jj] = analysis.compute_angles(W_0[jj]).min()
        model.components_ = W_0[jj]
        costs0[ii, jj] = model.losses(np.zeros((n_mixtures, 1)))[0]
```

```python
f = plt.figure(figsize=(5.5, 2))

le = .09
re = .03
be = .2
te = .075
hg = .075
h = 1. - be - te
w = (1. - le - re - hg) / 2.

ax0 = f.add_axes([le, be, w, h])
ax1 = f.add_axes([le + w + hg, be, w, h])
axes = [ax0, ax1]

normalized_costs = costs/costs0.mean(axis=1, keepdims=True)
normalized_costs -= normalized_costs.mean(axis=1, keepdims=True)
normalized_min_angles = (min_angles - min_angles.mean(axis=1, keepdims=True)) * np.pi / 180.
cost_bins = np.linspace(normalized_costs.min()*.99, normalized_costs.max()*1.01, 50)
angle_bins = np.linspace(normalized_min_angles.min()*.99, normalized_min_angles.max()*1.01, 50)
for ii, cost in enumerate(degeneracy_controls):
    label = styles.labels[cost]
    c = styles.colors.get(cost, 'b') 
    label = styles.labels.get(cost, cost)
    st = styles.line_styles.get(cost, '-') 
    pe_arg = [pe.Stroke(linewidth=styles.lw+1, foreground='k'), pe.Normal()]
    vals, _ = np.histogram(normalized_costs[ii], cost_bins, density=True)
    vals /= vals.sum()
    axes[0].plot(cost_bins[:-1] + np.diff(cost_bins)[0]/2., vals,
                 st, drawstyle=styles.ds, c=c, path_effects=pe_arg, label=label)
    vals, _ = np.histogram(normalized_min_angles[ii], angle_bins, density=True)
    vals /= vals.sum()
    axes[1].plot(angle_bins[:-1] + np.diff(angle_bins)[0]/2., vals, st, drawstyle=styles.ds, c=c, path_effects=pe_arg, label=label)
axes[0].set_ylabel('PDF', fontsize=styles.label_fontsize, labelpad=0)
axes[0].set_xlabel('Normalized Cost', fontsize=styles.label_fontsize)
axes[1].set_xlabel('Normalized Smallest Pairwise Angle (radians)', fontsize=styles.label_fontsize)
axes[0].set_xticks([-5e-4, 0, 5e-4])
axes[1].legend(loc='upper left', frameon=False, fontsize=styles.legend_fontsize,
              )
axes[0].set_xlim(cost_bins[0], cost_bins[-1])
axes[1].set_xlim(angle_bins[0], angle_bins[-1])
axes[1].set_xticks([-5e-2, 0, 5e-2])

f.text(0, .925, 'A', fontsize=styles.letter_fontsize)
f.text(w + hg + .05, .925, 'B', fontsize=styles.letter_fontsize)
for ax in axes:
    ax.set_ylim(1e-3, 2.)
    ax.set_yticks([1e-3, 1])
    ax.tick_params(labelsize=styles.ticklabel_fontsize)
    ax.set_yscale('log')
f.tight_layout()
plt.savefig(os.path.join(os.environ['HOME'], 'Downloads/sup9.pdf'), dpi=300)
plt.savefig(os.path.join(os.environ['HOME'], 'Downloads/sup9.png'), dpi=300)
```
